```python
station_names = [
    "Vienna",
    "Barbados",
]
test_dir = "/work/bm1159/XCES/xces-work/k203179/data/test"
```

```python
import xarray as xr
import matplotlib.pyplot as plt

for station_name in station_names:
    era5_for_station = xr.open_dataset(f"{test_dir}/era5_for_{station_name.lower()}.nc")
    
    for context in ["year", "intra_year", "intra_day"]:
        context_data = xr.open_dataset(f"{test_dir}/{context}_at_{station_name.lower()}.nc")

        plt.plot(context_data.time, context_data[context].values[:, 0, 0])
        plt.plot(context_data.time, context_data[context].values[:, 0, 7])
        plt.title(f"{context} at {station_name}")
        plt.show()

    plt.plot(era5_for_station.time, era5_for_station.tas.values.mean(axis=(1, 2)))
    plt.title
    plt.show()
```
